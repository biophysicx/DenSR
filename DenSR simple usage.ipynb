Use think chunk to test if DenSR is working, using inbuilt sample files

````python
from densr import densr, sample_data

sample = sample_data()

densr(spots1=sample.spots1,
    spots2=sample.spots2,
    object=sample.object,
    colocalization_threshold=0.25)

````

Use the following code to run DenSR with your own files.
Note: Replace the file paths and adjust colocalization thresholds

````python
from densr import densr

spots1 = "path/to/folder/C1_1a_xyz.txt"
spots2 = "path/to/folder/C1_1b_xyz.txt"
object = "path/to/folder/C1_volume.obj"
colocalization_threshold = 0.25

densr(spots1=spots1, 
      spots2=spots2, 
      object=object, 
      colocalization_threshold=colocalization_threshold)

````
